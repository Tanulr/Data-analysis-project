```python
#PES1UG20CS696 (Dataset-41.csv)

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/chiragr/Downloads/COVID_dataset.csv")
df
```

```python
#INTRODUCTORY QUESTIONS (IGNORING NULL VALUES)

#1. Create a column called fatality rate for each country Hint: Fatality rate= (fatalities/nó of cases)*100

fatality = []
cases = df['cases']
deaths = df['deaths']

for i in range(0, len(deaths)):
    if cases[i] == 0:
        fatality.append(0)
    else:
        fatality.append(deaths[i]/cases[i]*100)

df['fatality'] = fatality #creating new column
df
```

```python
#2. Plot the fatality rate of the country with the highest number of cases using a Bar Chart/Pie Chart

df['cases'].describe()
```

```python
quar_3 = 6.081192e+05 #finding 3rd quartile of cases

high_fatality = []
name = []
names = df['name']
for i in range(0, len(cases)):
    if cases[i] >= quar_3:
        high_fatality.append(fatality[i]) # appending the fatality rate and name of those countries which exceed 3rd quartile
        name.append(names[i])

high_fatrate = pd.DataFrame(list(zip(name, high_fatality)),columns =['name', 'fatality rate'])
plt.figure(figsize=(20,5))
sns.barplot(x = 'name',y = 'fatality rate', data=high_fatrate).set_title("Fatality Rate VS Cases")
```

```python
#3. Visualize the case count over the last 24 hours for each region in a bar plot.

plt.figure(figsize=(20,5))
sns.barplot(x = 'region',y = 'cases24h', data=df).set_title("Fatality Rate VS Cases")
```

```python
#4. Create a new column called population. The population is calculated by dividing the cases with cases100k and multiplying 
#   it by 100,000 for each country.


population = []
cases100k = df['cases100k']
cases = df['cases']

for i in range(0, len(cases)):
    if cases100k[i] == 0:
        population.append(0)
    else:
        population.append(cases[i]/cases100k[i]*100000)

df['population'] = population
df
```

```python
#TASK QUESTIONS
# 1.Filling null values. Identify the rows that have Null values in them. All of these null values can be filled using the 
#   existing data, Observe the relationship between the columns and use it to fill the Null columns. A little variation in your 
#   final answers is acceptable. Once you have filled these columns, fill in fatality rate and population for these rows as well
#   Hint: Use the population column.

df.isnull().sum()
```

```python
df['cases'].fillna(df['cases100k']*100000, inplace=True)
df['cases7days'].fillna(df['cases24h']*7,inplace=True)
df['cases7days100k'].fillna(df['cases7days']/df['population']*100000,inplace=True)

df['deaths'].fillna(df['deaths100k']*100000, inplace=True)
df['deaths100k'].fillna(df['deaths']/df['population']*100000,inplace=True)
df['deaths7days'].fillna(df['deaths24h']*7,inplace=True)
df['deaths7days100k'].fillna(df['deaths7days']/df['population']*100000,inplace=True)

df['fatality'].fillna(df['deaths']/df['cases']*100 ,inplace=True)
df['population'].fillna(df['cases']/df['cases100k']*100000 ,inplace=True)

df.isnull().sum()
```

```python
#3. Europe has been affected worse then the Americas by COVID. True or false? Justify with visualisations

#we takes cases100k as it gives proportional value, since total cases will be misleading as population if different
americas = 0
europe = 0
americas_cases100k = 0
americas_deaths100k = 0
europe_cases100k = 0
europe_deaths100k = 0

cases100k = df['cases100k']
deaths100k = df['deaths100k']
region = df['region']

for i in range(0, len(cases)):
    if region[i] == "Europe":
        europe +=  1
        europe_cases100k += cases100k[i]
        europe_deaths100k += deaths100k[i]
    if region[i] == "Americas":
        americas +=  1
        americas_cases100k += cases100k[i]
        americas_deaths100k += deaths100k[i]

europe_cases100k = europe_cases100k/europe
europe_deaths100k = europe_deaths100k/europe
americas_cases100k = americas_cases100k/americas
americas_deaths100k = americas_deaths100k/americas

print(europe_cases100k,europe_deaths100k,americas_cases100k,americas_deaths100k)
```
